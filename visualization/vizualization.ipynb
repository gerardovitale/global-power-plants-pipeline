```python
import pandas as pd
from sources.plottting import plot_pollutant, plot_stacked_pollutant
```

```python
spain_geo_air_path = '../outputs/spain_geo_air.csv'
spain_geo_air = pd.read_csv(spain_geo_air_path, index_col=0)
# spain_geo_air.head()
```

## Getting the mean pollutants per state
* aqi: Air Quality Index [US - EPA standard 0 - +500]
* pm10: Concentration of particulate matter < 10 microns (µg/m³)
* co: Concentration of carbon monoxide (µg/m³)
* o3: Concentration of surface O3 (µg/m³)
* so2: Concentration of surface SO2 (µg/m³)
* no2: Concentration of surface NO2 (µg/m³)

```python
pollution_per_state = spain_geo_air[
    ['aqi', 'pm10', 'co', 'o3', 'so2', 'no2']
].groupby(spain_geo_air.state).mean()

pollution_per_state.reset_index(level=0, inplace=True)

# pollution_per_state.head()
```

```python
plot_pollutant(pollution_per_state)
```

```python
plot_stacked_pollutant(pollution_per_state)
```
