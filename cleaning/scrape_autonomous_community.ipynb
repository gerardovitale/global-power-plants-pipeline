```python
import pandas as pd
from sources.get_autonomous_community import scrape_autonomous_community, clean_autonomous_community, clean_number_format
from sources.create_nested_list import create_nested_list
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Comunidades_y_ciudades_aut%C3%B3nomas_de_Espa%C3%B1a'
scrape_data = scrape_autonomous_community(url)
scrape_data = clean_autonomous_community(scrape_data)
scrape_data = clean_number_format(scrape_data)
scrape_data = create_nested_list(scrape_data, 9)
```

```python
column_names = ['id', 'name', 'capital', 'population', 
               'population_percent', 'population_density', 
               'area', 'density_percent', 'GDP']
```

```python
autonomous_community = pd.DataFrame(scrape_data[1:-1], columns=column_names)
autonomous_community.set_index('id', inplace=True)
```
